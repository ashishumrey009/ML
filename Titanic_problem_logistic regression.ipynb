```python
def age_approx(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            return 37
        if Pclass ==2:
            return 29
        if Pclass ==3:
            return 24
    else:
        return Age
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'] = titanic [['Age','Pclass']].apply(age_approx,axis=1)
```

```python
titanic.isnull().sum()
```

```python
titanic.drop(['Cabin'],axis=1,inplace=True)
```

```python
titanic.dropna(inplace=True,axis=0)
```
