```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
%matplotlib inline
```

```python
raw_data = loadmat("E:\MLNEW _BTM\svm\ex6data1.mat")
raw_data

data = pd.DataFrame(raw_data['X'],columns=['X1','X2'])
data['y'] = raw_data['y']
```

```python
from sklearn import svm
```

```python
svm = svm.LinearSVC(C=1,max_iter=1000)
```

```python
svm.fit(data[['X1','X2']],data['y'])
svm.score(data[['X1','X2']],data['y'])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
prediction = svm.predict(data[['X1','X2']])
```

```python
prediction
```

```python
print(confusion_matrix(prediction,data['y']))
```

```python
raw_data2 = loadmat("E:\MLNEW _BTM\svm\ex6data2.mat")
```

```python
raw_data2
```

```python

data = pd.DataFrame(raw_data2['X'],columns=['X1','X2'])
data['y'] = raw_data2['y']
```

```python
from sklearn import svm
svc = svm.SVC(C=100,gamma=0.3,probability = True)
```

```python
svc.fit(data[['X1','X2']],data['y'])
```

```python
svc.score(data[['X1','X2']],data['y'])
```

```python
parameters = {'kernel':('linear', 'rbf'), 'C':[0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100], 'gamma':[0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 50, 100]}
```

```python
svc=svm.SVC()
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svc,parameters,cv=5)
```

```python
clf.fit(data[['X1','X2']],data['y'])
```

```python
clf.best_params_
```

```python
clf.score(data[['X1','X2']],data['y'])
```
